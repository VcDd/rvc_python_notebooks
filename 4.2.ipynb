# Lecture 4.2

Polynomial trajectories

```python
#Create a polynomial trajectory using the toolbox function tpoly(), 
#passing in the initial position of 0 and a final position of 1 over 50 timesteps:
#The function will return the vector s,sd, and sdd
data=tr.tpoly(0, 1, 50)
t=data.x
s=data.y
sd=data.yd
sdd=data.ydd
print(s,sd,sdd)

#plot
fig1=plt.figure(1)
#1
sfig1=fig1.add_subplot(3,1,1)
plt.plot(t,s)
#2
sfig2=fig1.add_subplot(3,1,2)
plt.plot(t,sd)
#3
sfig3=fig1.add_subplot(3,1,3)
plt.plot(t,sdd)
```

```python
data=tr.tpoly(0, -1, 50)
t=data.x
s=data.y
sd=data.yd
sdd=data.ydd
#plot
fig2=plt.figure(2)
#1
sfig1=fig2.add_subplot(3,1,1)
plt.plot(t,s)
#2
sfig2=fig2.add_subplot(3,1,2)
plt.plot(t,sd)
#3
sfig3=fig2.add_subplot(3,1,3)
plt.plot(t,sdd)
```

```python
#The initial and final velocities can be explicitly specified by passing in additional arguments
#in this case 0.5 and 0, respectively:
data=tr.tpoly(0, 1, 50, 0.5, 0)
t=data.x
s=data.y
sd=data.yd
sdd=data.ydd

#plot
fig3=plt.figure(3)
#1
sfig1=fig3.add_subplot(3,1,1)
plt.plot(t,s)
#2
sfig2=fig3.add_subplot(3,1,2)
plt.plot(t,sd)
#3
sfig3=fig3.add_subplot(3,1,3)
plt.plot(t,sdd)
```
