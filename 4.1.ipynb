Create a **final pose** comprised of an offset of:
 1 unit in the x direction 
 2 units in the y direction
 3 units in the z direction  

And an orientation described by RPY angles of (0.6,0.8,1.4) radians

```python
T1=transl(1,2,3).dot(rpy2tr(0.6,0.8,1.4))
print(T1)
```
