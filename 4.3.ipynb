# Lecture 4.3

1D trapezoidal trajectories

```python
#Create a trapezoidal trajectory using the toolbox function lspb(),
#passing in the initial position of 0 and a final position of 1 over 50 timesteps
#The function will return the vector s,sd, and sdd
data=tr.lspb(0, 1, 50)
t=data.x
s=data.y
sd=data.yd
sdd=data.ydd
```

```python
#plot;
fig1=plt.figure(1);
#1
sfig1=fig1.add_subplot(3,1,1);
plt.plot(t,s);
#2
sfig2=fig1.add_subplot(3,1,2);
plt.plot(t,sd);
#3
sfig3=fig1.add_subplot(3,1,3);
plt.plot(t,sdd);

#print(s,sd,sdd)
```

```python
#Changing the value of the final position from 1 to -1 gives
data=tr.lspb(0, -1, 50)
t=data.x
s=data.y
sd=data.yd
sdd=data.ydd
#plot
fig2=plt.figure(2);
#1
sfig1=fig2.add_subplot(3,1,1);
plt.plot(t,s);
#2
sfig2=fig2.add_subplot(3,1,2);
plt.plot(t,sd);
#3
sfig3=fig2.add_subplot(3,1,3);
plt.plot(t,sdd);
```

```python
#The top velocity of the linear segment can be explicitly specified by passing in additional argument, in this case 0.025:
data=tr.lspb(0, 1, 50, 0.025)
t=data.x
s=data.y
sd=data.yd
sdd=data.ydd
```

```python
#plot
fig3=plt.figure(3);
#1
sfig1=fig3.add_subplot(3,1,1);
plt.plot(t,s);
#2
sfig2=fig3.add_subplot(3,1,2);
plt.plot(t,sd);
#3
sfig3=fig3.add_subplot(3,1,3);
plt.plot(t,sdd);
```

```python
#Increasing the velocity of the linear segment to 0.035 gives:
data=tr.lspb(0, 1, 50, 0.035)
t=data.x
s=data.y
sd=data.yd
sdd=data.ydd

#plot
fig4=plt.figure(4)
#1
sfig1=fig4.add_subplot(3,1,1)
plt.plot(t,s)
#2
sfig2=fig4.add_subplot(3,1,2)
plt.plot(t,sd)
#3
sfig3=fig4.add_subplot(3,1,3)
plt.plot(t,sdd)
```
